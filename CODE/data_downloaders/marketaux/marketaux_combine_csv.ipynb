```python
import pandas as pd
import os


df_tmp = pd.read_csv("C:/Users/szymo/LBN-INZ/DATA/marketaux/news/BTCUSD/BTCUSD_20210101_20210104.csv")
df = pd.DataFrame(columns=df_tmp.columns)
# Create one df from all dfs in this directory
for filename in os.listdir('C:/Users/szymo/LBN-INZ/DATA/marketaux/news/BTCUSD'):
    if filename.endswith(".csv"):
        df_tmp = pd.read_csv('C:/Users/szymo/LBN-INZ/DATA/marketaux/news/BTCUSD/' + filename)
        df = pd.concat([df, df_tmp], ignore_index=True)

df
```

```python
df.to_csv('C:/Users/szymo/LBN-INZ/DATA/marketaux/news/BTCUSD_20210101_20231126.csv', index=False)
```
